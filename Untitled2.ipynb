```python
import pymongo
from google.colab import userdata

# Conecta usando seu Secret
uri = userdata.get('MONGO_URI')
client = pymongo.MongoClient(uri)
db = client.Projeto_H

# Criando uma coleção chamada 'usuarios' e inserindo o primeiro registro
colecao = db.usuarios
dados_iniciais = {
    "nome": "Pablo Leonardo",
    "funcao": "Admin/Dev",
    "status": "Online",
    "projeto": "Miranda"
}

# Insere no banco
resultado = colecao.insert_one(dados_iniciais)

print(f"✅ Documento criado com ID: {resultado.inserted_id}")
print(f"📂 Coleções agora no banco: {db.list_collection_names()}")
```
